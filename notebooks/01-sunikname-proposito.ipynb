```python
import numpy as np
```

```python
casas_entradas = np.load('../data/entradas.npy')
casas_salidas = np.load('../data/salida_precios.npy')
```

```python
print(casas_entradas[:5])
```

```python
print(casas_salidas[:5])
```

```python
min_vals = casas_entradas.min(axis=0)
max_vals = casas_entradas.max(axis=0)
casas_entradas_normalizadas = (casas_entradas-min_vals)/(max_vals - min_vals)

min_vals = casas_salidas.min(axis=0)
max_vals = casas_salidas.max(axis=0)
casas_salidas_normalizadas = (casas_salidas-min_vals)/(max_vals - min_vals)
```

```python
import sys
sys.path.append('../libs')
import neuronas
import importlib
importlib.reload(neuronas)
from neuronas import MLP    
from neuronas import Plotter   
```

```python
RNA_MLP =  MLP([4,10,1])
```

```python
predicciones_precios_casas = RNA_MLP.feedforward(casas_entradas_normalizadas)
```

```python
predicciones = predicciones_precios_casas * (max_vals - min_vals) + min_vals
Ys = np.stack((predicciones, casas_salidas, casas_salidas - predicciones)).squeeze().T
```

```python
RNA_MLP_10HL_a = MLP([4,10,1])
```

```python
predicciones_precios_casas = RNA_MLP_10HL_a.feedforward(casas_entradas_normalizadas)
predicciones = predicciones_precios_casas * (max_vals-min_vals) + min_vals
```

```python
Ys2 = np.stack((predicciones, casas_salidas, casas_salidas - predicciones)).squeeze().T
```
